```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
import csv
import seaborn as sns

from konlpy.tag import Okt
from konlpy.tag import Mecab
```

```python
df = pd.read_csv("./crawling/IT_crawling_솔직 후기.csv", index_col=0)
```

```python
df.reset_index(inplace=True)
```

```python
for i in range(len(df['content'])):
    df['content'][i] = df['content'][i].replace('\n', ' ')
    
```

```python
df['content']
```

# content 글자 수

```python
df['content_count'] = 0
```

```python
for i in range(len(df['content'])):
    df['content_count'][i] = len(df['content'][i])
```

# Content

```python
okt = Okt()
```

```python
cleaned_sentences = pd.Series()
clean_content = []

for sentence in df['content']:
    
    sentence = re.sub('[^가-힣|A-Z|+|-]', ' ', sentence)
    
    token= okt.pos(sentence, norm=True, stem=True) #전에는 mos를 써는데 pos를 쓰면 튜플로(쌍으로) ('','noun')형태로 만들어 준다. 
    
    df_token = pd.DataFrame(token, columns=['word', 'class'])
    
    cleaned_df_token = df_token[(df_token['class']=='Noun')|(df_token['class']=='Verb')|(df_token['class']=='Adjective')]
    
    
    cleaned_sentences = cleaned_df_token['word']
    cleaned_sentences = ' '.join(cleaned_sentences)
    clean_content.append(cleaned_sentences)
    print(cleaned_sentences)
    

df['cleaned_content'] = clean_content
```

# cleaned_content 글자수

```python
df['cleaned_content_count'] = 0
```

```python
for i in range(len(df['cleaned_content'])):
    df['cleaned_content_count'][i] = len(df['cleaned_content'][i])
```

```python
cleaned_token_review = list(df['cleaned_content'])
cleaned_sentences = []

for sentence in cleaned_token_review:
    token = sentence.split(' ')
    cleaned_sentences.append(token)

df['cleaned_content'] = cleaned_sentences
```

# 블루투스 단어수 

```python
df['bluetooth_word_count'] = 0

for j in range(len(df['cleaned_content'])):
    for i in df['cleaned_content'][j]:
        if i == '블루투스':
            df['bluetooth_word_count'][j] += 1
```

# stopword을 cleaned_content에 적용

```python
stopwords = pd.read_csv('./datasets/stopwords_for_blog.csv', index_col=0)
stopwords.head()
```

```python
cleaned_words = []
    
for i in range(len(df['cleaned_content'])):

    words = []

    for word in df['cleaned_content'][i]:
        if i >= 0:
            if word not in (list(stopwords['stopword'])):
                words.append(word)

            else:
                print(word)
        else:
            print(word)
    

    cleaned_word = ' '.join(words)
    cleaned_words.append(cleaned_word)
    
df['very_cleaned_content'] = cleaned_words
```
